```python
import re

# Create function to fix dates
def fix_date_format(date_str):
    # Check 4th character, if '-' remove it, otherwise return original string
    return re.sub(r'^(...)-', r'\1', date_str)

def change_to_date(df, date_cols):
    
    # Loop through list to change column dtype to datetime
    for col in date_cols:
        df[col] = pd.to_datetime(df[col]).astype("datetime64[D]")

# Function to replace leap year dates
def fix_leap_year_dates(df, columns):
    for col in columns:
        df.loc[:, col] = df[col].str.replace(r"(\d{4})-02-29", r"\1-02-28", regex=True)
```

```python
# List of date columns to clean
contracts_date_columns = ["Application_Date"]

for col in contracts_date_columns:
    application_df[col] = application_df[col].astype(str).apply(fix_date_format)

for col in contracts_date_columns:
    application_df[col] = pd.to_datetime(application_df[col], errors='coerce')

application_df.head()
```

```python
# Replace NaN in object (string) columns with "Unknown"
application_df.loc[:, application_df.dtypes == "object"] = application_df.loc[:, application_df.dtypes == "object"].fillna("Unknown")

# Replace NaN in float64 columns with the mean of the respective column
application_df.loc[:, application_df.dtypes == "float64"] = application_df.loc[:, application_df.dtypes == "float64"].apply(lambda x: x.fillna(x.mean()))

# Verify the changes
application_df.isnull().sum()  # Check if any nulls remain
```

```python
contract_snapshot_df["Number of Days Past Due"].fillna(0, inplace=True)
contract_snapshot_df.isna().sum()
```

```python

contractSnapshot_date_cols = ["Snapshot Date"]

# Apply the function to fix the date column
for col in contractSnapshot_date_cols:
    contract_snapshot_df[col] = contract_snapshot_df[col].apply(fix_date_format)

# Fix Leap year
fix_leap_year_dates(contract_snapshot_df, contractSnapshot_date_cols)

# Change to datetime datatype
change_to_date(contract_snapshot_df, contractSnapshot_date_cols)
# Convert 'Snapshot Date' and 'Actual_Contract_End_Date' to datetime format



# Get the latest Snapshot Date for each Contract_ID
# Keep only the latest snapshot for each Contract_ID
latest_snapshots = contract_snapshot_df.sort_values(by=["Contract_ID", "Snapshot Date"], ascending=[True, False]).drop_duplicates(subset=["Contract_ID"], keep="first")

latest_snapshots.head()


```

```python
latest_snapshots.dropna(subset=["Snapshot Date"], inplace=True)
latest_snapshots.head()
```

```python
contract_date_cols = ["Actual_Contract_End_Date"]

# Safe function to clean date format
def fix_actual_date_format(date_str):
    if pd.isna(date_str):  # Check if value is NaN or None
        return date_str  # Return as is (to keep it as NaN)
    if isinstance(date_str, str):  # Only process strings
        return re.sub(r'^(...)-', r'\1', date_str)
    return date_str  # If it's already a datetime, leave it



# Apply the function to fix the date column
for col in contract_date_cols:
    contract_data[col] = contract_data[col].apply(fix_actual_date_format)

# Fix Leap year
fix_leap_year_dates(contract_data, contract_date_cols)

# Change to datetime datatype
change_to_date(contract_data, contract_date_cols)



# Merge the latest snapshots with ContractData based on 'Contract_ID'
contract_data = contract_data.merge(latest_snapshots[["Contract_ID", "Snapshot Date"]], on="Contract_ID", how="left")
contract_data.head()
```

```python
# Fill null values in ActualContract_End_Date with values from latest_snapshots
contract_data["Actual_Contract_End_Date"].fillna(contract_data["Snapshot Date"], inplace=True)

# Drop the extra column after filling (optional)
contract_data.drop(columns=["Snapshot Date"], inplace=True)

# Display the updated DataFrame
contract_data.head()
```

```python
current_saving_df = pd.read_csv('CurrentandSavingsAccountData.csv')
current_saving_df.head()
```

```python
current_saving_df = current_saving_df.loc[:, current_saving_df.isnull().mean() <= 0.5]
current_saving_df.info()
```

```python
current_saving_df.dropna(subset=['SpshotMonth'], inplace=True)
```

```python
current_saving_df.info()
```

```python
current_saving_df.isna().sum()
```

```python
curr_savings_date_cols = ["Last Transaction Date"]

# Fill missing dates
for col in curr_savings_date_cols:
    current_saving_df[col].fillna("1900-01-01", inplace=True)

# Identify numerical columns
numerical_cols = current_saving_df.select_dtypes(include=["number"]).columns

# Fill null values with 0 in numerical columns
current_saving_df[numerical_cols] = current_saving_df[numerical_cols].fillna(0)
    

# Confirm
current_saving_df.isna().sum()
```

```python
current_saving_df.info()
```

```python
current_saving_date_cols = ["Month (YYYY-MM)","Open Date","Last Transaction Date","SpshotMonth"]

for col in current_saving_date_cols:
    current_saving_df[col] = current_saving_df[col].apply(fix_date_format)

# Fix Leap year
fix_leap_year_dates(current_saving_df, current_saving_date_cols)
                    

# Change to datetime datatype
change_to_date(current_saving_df, current_saving_date_cols)

# Confirm 
current_saving_df.info()
```

```python
# Convert to last day of the previous month
current_saving_df["Snapshot Date"] = current_saving_df["Month (YYYY-MM)"] - pd.DateOffset(days=1)
current_saving_df.head()
```
