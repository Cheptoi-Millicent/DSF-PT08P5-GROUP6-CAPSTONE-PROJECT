```python
contract_snapshot_df = pd.read_csv('ContractsSnapshotData.csv')
contract_snapshot_df.head()
```

```python
import re


# Function to fix incorrect year formatting
def fix_date_format(date_str):
    if pd.isna(date_str):  # If the value is NaN, return it as is
        return None
    
    # Match incorrect year format (e.g., "202-1-08-31")
    match = re.match(r"(\d{3})-(\d)-(\d{2})-(\d{2})", date_str)
    if match:
        corrected_year = match.group(1) + match.group(2)  # Fix year part
        corrected_date = f"{corrected_year}-{match.group(3)}-{match.group(4)}"
        return corrected_date
    
    return date_str  # Return original date if it's correctly formatted

# List of date columns to clean in both datasets
contracts_date_columns = ["Contract_Start_Date", "Expected_Contract_End_Date", "Actual_Contract_End_Date"]
snapshot_date_columns = ["Snapshot Date"]

# Fix date formats in ContractsData
for col in contracts_date_columns:
    contract_df[col] = contract_df[col].astype(str).apply(fix_date_format)

# Fix date format in ContractsSnapshotData
for col in snapshot_date_columns:
    contract_snapshot_df[col] = contract_snapshot_df[col].astype(str).apply(fix_date_format)

# Convert to datetime format after fixing
for col in contracts_date_columns:
    contract_df[col] = pd.to_datetime(contract_df[col], errors='coerce')

for col in snapshot_date_columns:
    contract_snapshot_df[col] = pd.to_datetime(contract_snapshot_df[col], errors='coerce')
```
