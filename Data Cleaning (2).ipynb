## **Merge Datasets: Application_Data AND CRB_Data**

#### The first merge being the application_data and the CRB_Data, then based on the output given we pick the required columns resuling to merge1

```python
import pandas as pd

# Merge application_data with CRB_Data
merged_df = pd.merge(application_data, CRB_Data, how='inner')

# Display the first few rows of the merged dataset
merged_df.head()
```

```python
## The necessary columns from the merged_df that comprises of application_data and CRB_Data
merge1_df = merged_df.loc[:, ['Loan_Approval_Status', 'Employment_Status','Time_at_Current_Employment_(Months)', 'Number_of_Dependents','Monthly_Income','CRB Score', 'CRB Grade']]
merge1_df.head()
```

## **Concationation: Current and Savings Account Data**

#### Appending Columns (Side-by-Side Merge), If you want to combine different columns from multiple DataFrames without a common key, use pd.concat() with axis=1:

```python

# Select specific columns to concat
merged_df2 = pd.concat([account_data[['Average Balance', 'Overdraft Limit', 'Sum of Monthly Credit Transactions', 'Number of Bounced Cheques']],
                       snapshot_data[['Outstanding Amount', 'Number of Days Past Due', 'Contract Status']]], axis=1)


merged_df2.head()
```

## **ContractsData: Picked the necessary columns from the data**

```python
# Selecting only the specified columns from the ContractsData since the application_data and Contracts id is different from the rest of the data
contracts_df = contracts_data[['Loan_Type', 'Loan_Term', 'Total_Loan_Amount']]

# Displaying the first few rows
contracts_df.head()
```

## **Data Cleaning**

```python
# Checking the columns' titles and datatypes
combined_df.info()
```

```python
# Get missing values count and percentage

missing_values = combined_df.isnull().sum()

missing_percentage = (combined_df.isnull().sum() / len(combined_df)) * 100

# Create a summary dataframe

missing_summary_df = pd.DataFrame({
    'Missing Values': missing_values,
    'Percentage Missing': missing_percentage})

# Sort by percentage missing
missing_summary = missing_summary_df.sort_values('Percentage Missing', ascending=False).round(0);

missing_summary
```

```python


# Fill categorical columns with mode (most frequent value)
categorical_cols =combined_df.select_dtypes(include=['object']).columns
for col in categorical_cols:
    combined_df[col].fillna(combined_df[col].mode()[0], inplace=True)

# Fill numerical columns with median (better for skewed data)
numerical_cols = combined_df.select_dtypes(include=['number']).columns
for col in numerical_cols:
    combined_df[col].fillna(combined_df[col].median(), inplace=True)

# Check if there are any remaining missing values
remaining_missing = combined_df.isnull().sum().sum()

print("Remaining Missing Values:", remaining_missing)
```

```python
crb = combined_df.isnull().sum()
crb
```
